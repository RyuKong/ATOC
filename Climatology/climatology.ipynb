```python
## %matplotlib inline
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import numpy as np
import statistics
import calendar



class WD:
    long = 0
    lat = 1
    jan = 2
    feb = 3
    mar = 4
    apr = 5
    may = 6
    jun = 7
    jul = 8
    aug = 9
    sep = 10
    oct = 11
    nov = 12
    dec = 13
    annual_min = 14
    annual_max = 15
    annual_avg = 16
    annual_sum = 17
  
    
    
def selection(total_list,choice,months,year):
    air_temp_data = total_list[0]
    precip_data = total_list[1]
    if choice == "1":
        plot(air_temp_data,months,year)
        plot(precip_data,months,year)
        
    elif choice == "2":
        print(" ")
        binsizes = int(input("How many bins do you need? (example: 10)"))
        hist(air_temp_data,binsizes,year)
        hist(precip_data,binsizes,year)
    
    elif choice == "3":
        bargraph(air_temp_data,year,months)
        bargraph(precip_data,year,months)
        
    elif choice == "4":
        heatmap(air_temp_data,year)
        heatmap(precip_data,year)
        
    elif choice == "5":
        aridity(air_temp_data,precip_data,year)
        
    elif choice == "6":
        boxplot(air_temp_data,year)
        boxplot(precip_data,year)

    elif choice == "7":
        piechart(precip_data,months,year)
        
        
        
def piechart(values, months, year): # after finishing up the required tasks
    index = 2                       # we continued and created a piechart graph
    lists = []
    while index < len(values[0][:WD.annual_min]):
        lists.append(statistics.mean(get_column_values(values,index)))
        index += 1
    
    plt.figure(figsize=(5,5))
    plt.pie(lists, labels=months, autopct='%1.1f%%')
    plt.title("Piechart of average precipitation per month in year " + year)
    plt.savefig("Piechart-precip-per-month" + year + ".pdf", bbox_inches='tight')
    plt.axis("image")
    plt.show()
        
        
        
def boxplot(values, year):
    delhi = []      # 28.75N, 77.25E
    losangeles = [] # 34.25N, 118.25W
    newyork = []    # 40.75N, 74.25W
    shanghai = []   # 31.25N, 121.25E
    tokyo = []      # 35.75N, 139.75E
    for each in values:
        if each[WD.long]==77.25 and each[WD.lat]==28.75:
            delhi.extend(each[WD.jan:WD.annual_min])
        if each[WD.long]==-118.25 and each[WD.lat]==34.25:
            losangeles.extend(each[WD.jan:WD.annual_min])
        if each[WD.long]==-74.25 and each[WD.lat]==40.75:
            newyork.extend(each[WD.jan:WD.annual_min])
        if each[WD.long]==121.25 and each[WD.lat]==31.25:
            shanghai.extend(each[WD.jan:WD.annual_min])
        if each[WD.long]==139.75 and each[WD.lat]==35.75:
            tokyo.extend(each[WD.jan:WD.annual_min])
            
    names = ['Delhi', 'Los Angeles', 'New York', 'Shanghai', 'Tokyo']
    data = [delhi,losangeles,newyork,shanghai,tokyo]
    plt.figure(figsize=(8,4))
    ax = plt.subplot()
    plt.xlabel("Five Megacities")
    if values[0][WD.annual_min] < 0:
        plt.title("Boxplot comparisons of temperature in " + year)
        plt.ylabel("Temperature (in degree Celsius)")
        ax.boxplot(data, labels=names)
        plt.savefig("Boxplot-airtemp" + year + ".pdf", bbox_inches='tight')
    else:
        plt.title("Boxplot comparisons of precipitation in " + year)
        plt.ylabel("Precipitation (in millimeter)")
        ax.boxplot(data, labels=names)
        plt.savefig("Boxplot-precip-" + year + ".pdf", bbox_inches='tight')
    plt.show()
    
    
    
def aridity(airtemp, precip, year): # http://www.fao.org/nr/climpag/globgrids/KC_commondata_en.asp
    index = 0                       # In the article it provides an simple aridity formula: 
    lat = []                        # Aridity = (Average precipitation/(Average temperature + 10)) while Average temperature of the coordinate is >= -9.9
    long = []                       # The formula is not suitable to calculate the aridity of cold zones.
    aridity = []                    # Because the calculations based on data from cold zones would led to unusual results 
    for each in airtemp or precip:  # So there are some areas shown grey in the graph, but all showed normal compared to the Global Aridity Map
        long.append(each[WD.long])
        lat.append(each[WD.lat])
    while index < len(airtemp):
        airtemp_avg = airtemp[index][WD.annual_avg] # make each individual of airtemp 
        precip_avg = precip[index][WD.annual_sum]   # corresponds to data from precipitation
        if (airtemp_avg < -9.9): # Daterminating cold zones
            aridity_avg = 100
        else:
            aridity_avg = precip_avg/(airtemp_avg + 10) # Aridity = (Average precipitation/(Average temperature + 10))
        aridity.append(aridity_avg)
        index += 1
    plt.figure(figsize=(10,5))
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    cm = matplotlib.cm.get_cmap('nipy_spectral')
    plt.title("Heat map Aridity Index values in year " + year)
    plt.scatter(long, lat, c=aridity, cmap=cm, vmin=0, vmax=100, s=1)
    plt.colorbar()
    plt.savefig("Global-aridity-" + year + ".pdf", bbox_inches='tight')
    plt.show()
        
        
    
def heatmap(values, year):
    lat = []
    long = []
    total = []
    for each in values: # to get all coordinates
        long.append(each[WD.long])
        lat.append(each[WD.lat])
        if values[0][WD.annual_min] < 0:      # if annual minimum is less than 0 s
            total.append(each[WD.annual_avg]) # it means the data is from air temp list
        else:
            total.append(each[WD.annual_sum])
    plt.figure(figsize=(10,5))
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    cm = matplotlib.cm.get_cmap('viridis')  
    if values[0][WD.annual_min] < 0:
        plt.title("Heat map temperature values in year " + year)
        plt.scatter(long, lat, c=total, cmap=cm, s=1)
        plt.colorbar()
        plt.savefig("Heatmap-airtemp-" + year + ".jpg", bbox_inches='tight')
    else:
        plt.title("Heat map precipitation values in year " + year)
        plt.scatter(long, lat, c=total, cmap=cm, norm=colors.LogNorm(), s=1)
        plt.colorbar()
        plt.savefig("Heatmap-precip-" + year + ".jpg", bbox_inches='tight')
    plt.show()


        
def bargraph(values, year, months):
    index = 2
    max_lst = []
    nnine_lst = []
    median_lst = []
    plt.figure(figsize=(8,4))
    while index < len(values[0][:WD.annual_min]): # the index starts at 2 and use list slicing
        lst = get_column_values(values,index)     # to get the range from jan to dec and calculate 
        percentile = np.percentile(lst, 99)       # different percentile by np.percentile and statistics
        nnine_lst.append(percentile)
        median = statistics.median(lst)
        median_lst.append(median)
        maximum = max(lst)
        max_lst.append(maximum)
        index += 1
    plt.xlabel("Month")
    plt.bar(months, max_lst, label='Maximum')
    plt.bar(months, nnine_lst, label='99th percentile')
    plt.bar(months, median_lst, label='50th percentile')
    plt.legend(loc=2, bbox_to_anchor=(1.02, 0.65))
    if median_lst[0] < 0:
        plt.ylabel("Temperature (in degree Celsius)")
        plt.title("Temperature observations per month in year " + year)
        plt.savefig("Bar-airtemp-" + year + ".pdf", bbox_inches='tight')
    else:
        plt.ylabel("Precipitation (in millimeter)")
        plt.title("Precipitation observations per month in year " + year)
        plt.savefig("Bar-precip-" + year + ".pdf", bbox_inches='tight')
    plt.show()
    
    
    
def hist(values, binsizes, year):
    legend = ['maximums', 'minimums'] 
    min_lst = get_column_values(values, WD.annual_min) # collect minimum data
    max_lst = get_column_values(values, WD.annual_max) # collect maximum data
    plt.figure(figsize=(8,4))
    plt.ylabel("Number of locations") 
    if min(min_lst) < 0: # if the minimum < 0 then it would be from temp data
        plt.xlabel("Temperature (in Celsius)") 
        plt.title("Histogram of max/min global temperature in " + year)
        plt.hist([max_lst,min_lst], bins = binsizes)
        plt.legend(legend)
        plt.savefig(str(binsizes) + "-Hist-airtemp-" + year + ".pdf", bbox_inches='tight')
    else: # else it would be from precip data
        plt.xlabel("Precipitation (in millimeter)") 
        plt.title("Histogram of max/min global precipitation in " + year) 
        plt.hist([max_lst,min_lst], bins = binsizes)   
        plt.legend(legend)
        plt.savefig(str(binsizes) + "-Hist-precip-" + year + ".pdf", bbox_inches='tight')
    plt.show() 
    
    
    
def plot(values, months, year):
    plt.figure(figsize=(8,4))
    ac = get_located_between(values,WD.lat,23.5,90) # Above the tropic of cancer
    bc = get_located_between(values,WD.lat,-90,-23.5) # Below the tropic of capricorn
    be = get_located_between(values,WD.lat,-23.5,23.5) # Between two tropic lines
    plt.plot(months, average(ac), label='above tropic of cancer')   # label months as xs and get average of the  
    plt.plot(months, average(be), label='between both tropics')     # specific months by average function and while loop
    plt.plot(months, average(bc), label='below tropic of capricorn')
    plt.legend(loc=2, bbox_to_anchor=(1.02, 0.65))
    
    if average(bc)[6] < 0:
        plt.xlabel("Month") 
        plt.ylabel("Temperature (in degree Celsius)") 
        plt.title("Line graph of global mean temperatures values over 12 months in year " + year) 
        plt.savefig("Plot-airtemp-" + year + ".pdf", bbox_inches='tight')
    else:
        plt.xlabel("Month") 
        plt.ylabel("Precipitation (in millimeter)") 
        plt.title("Line graph of global mean precipitation values over 12 months in year " + year) 
        plt.savefig("Plot-precip-" + year + ".pdf", bbox_inches='tight')
    plt.show()
        
    
    
def average(values): # This function is to get average result of lists
    column = 0
    average = []
    while column < len(values[0]):
        avg_temp_ac = get_column_values(values,column)
        avg = statistics.mean(avg_temp_ac)
        average.append(round(avg,2))
        column += 1
    return average



def get_menu_choice():
    print(" ")
    print("1: Line graph of global mean temperature/precipitation over 12 months.")
    print ("2: Histogram of minimum/maximum temperatures at all locations for a given year.")
    print ("3: Temperature/precipitation over month in percentile.")
    print ("4: Heat map of mean annual temperature/precipitation.")
    print ("5: Heat map of simple aridity index of De Martonne.")
    print ("6: Boxplots comparisons of temperature/precipitation of 5 megacities.")
    print ("7: Piechart of verage precipitation observations per month.")
    print ("8: Select another year.")
    print("0: Exit")
    choice = input("> ")
    return choice



def verify_climatology_data(total):
    data_ind = 0
    while data_ind < len(total):
        data = total[data_ind]
        data_ind += 1
        rows = len(data)
        cols = len(data[0])
        same_number_columns = True
        i = 0
        while i < len(data):
            row = data[i]
            if len(row) != cols:
                same_number_columns = False
            i += 1
        if not same_number_columns:
            print("Some rows have a different number of columns")
        overall_type = True
        column = 0
        while column < cols:
            col_vals = get_column_values(data, column)
            correct_type = True
            j = 0
            while j < len(col_vals):
                val = col_vals[j]
                if type(val) is not float:
                    correct_type = False
                j += 1
            if not correct_type:
                print("column " + str(column) + " contains values that are not the right type!")
                print(col_vals[:10])
                overall_type = False
            column += 1

            if overall_type and same_number_columns:
                if rows == 85794 and cols == 18:
                    return 
                else:
                    print(" ")
                    print("You have the incorrect shape of data!")
                    print("You have # rows (should be 85794): " + str(rows))
                    print("You have # columns (should be 18): " + str(cols))
                    return
                
                
                
def get_located_between(values,index,minimum,maximum):
    records = []
    for each in values:
        if minimum <= each[index] < maximum:
            records.append(each[WD.jan:WD.annual_min]) # If the each[index] fits the interval 
    return records                                     # then append the lists to 'records' and return them 



def get_column_values(data, column):
    col_vals = []
    for each in data:
        col_vals.append(each[column])
    return col_vals



def get_month_names():
    months = []
    months.extend(calendar.month_abbr[1:13])
    return months
    


def read_data(filename):  # this is the function that reads the file and
    index = 0             # read the original data from files
    total_list = []       # converts all elements to integers in the 
    air_temp_list = []    # list of lists before returning it
    precip_list = []
    while index < len(filename):
        file = open(filename[index], "r")
        lines = file.readline()
        lists = []
        while lines:
            floats = []
            data = lines.split()
            for one in data:
                floats.append(float(one))
            lists.append(floats)
            lines = file.readline()
        for each in lists:
            month = each[WD.jan:]
            each.append(min(month))
            each.append(max(month))
            each.append(round(sum(month)/12,2))
            each.append(round(sum(month),2))
            if index == 0:
                air_temp_list.append(each)
            elif index == 1:
                precip_list.append(each)
        file.close()
        index += 1
    total_list.append(air_temp_list) # making the air temperature data and precipitation data to a total list
    total_list.append(precip_list)   # therefore it only needs the read_data function to run one time
    return total_list

    
    
def ask_year():
    print("Please enter the year you would like to select between 1900 & 2014 (example: 1947)")
    year = input("> ")
    return year
    
    
    
def main():
    year = ask_year()
    if year.isnumeric() == True:
        filename = [("air_temp." + year),("precip." + year)]
        if int(year) >= 1900 and int(year) <= 2014:
            print(" ")
            print("Thank you for selecting the year " + year)
            months = get_month_names()
            total_list = read_data(filename)
            verify_climatology_data(total_list)
            while True:
                choice = get_menu_choice()
                if choice == "0":
                    break
                elif choice == "1" or choice == "2" or choice == "3" or choice == "4" or choice == "5" or choice == "6" or choice == "7":
                    selection(total_list,choice,months,year)
                elif choice == "8":
                    print(" ")
                    main()
                    break      
        else:
            print(" ")
            print("The year " + year + " does not exist in database!")
    else:
        print(" ")
        print("Retry and enter a valid input!")
        print(" ")
        main()
if __name__ == "__main__":
    main()
```
